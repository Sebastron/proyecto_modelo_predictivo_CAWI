```python
dias_semana = [['Lunes', 656526], ['Martes', 735993], ['Miercoles', 788368], ['Jueves', 785624], ['Viernes', 679215], ['Sabado', 562904], ['Domingo', 384976]]
d_s = pd.DataFrame(dias_semana, columns=['dia_semana','Cantidad_dia'])
d_s['Probabilidad_dia'] = d_s['Cantidad_dia']/d_s['Cantidad_dia'].sum()
d_s['Cantidad_dia'].sum()
```

```python
horas = [['8', 281809], ['9', 332302], ['10', 361548], ['11', 382929], ['12', 320620], ['13', 356156], ['14', 325255], ['15', 338795], ['16', 343691], ['17', 373910], ['18', 332011], ['19', 354036], ['20', 324060], ['21', 166484]]
d_h = pd.DataFrame(horas, columns=['Hora','Cantidad_hora'])
d_h['Probabilidad_hora'] = d_h['Cantidad_hora']/d_h['Cantidad_hora'].sum()
d_h['Probabilidad_hora'].sum()
```

```python
array = []
for i in range (len(d_s['dia_semana'])):
  for j in range (len(d_h['Hora'])):
    probabilidad = d_s['Probabilidad_dia'][i]*d_h['Probabilidad_hora'][j]
    pro = round(probabilidad, 4)
    array.append([d_s['dia_semana'][i], d_h['Hora'][j], pro])
df = pd.DataFrame(array, columns=['Dia_semana','Hora', 'Probabilidad_respuesta'])
df
```

```python
def Prediccion_horario(cantidad_respuestas):
  horario = pd.read_csv(path + 'Probabilidad_horario_general.csv', sep=';')
  horario["Cantidad_estimada"] = (horario["Probabilidad_respuesta"]*cantidad_respuestas).round(0).astype(int)
  horario = horario.drop('Probabilidad_respuesta', axis=1)
  return horario
x = Prediccion_horario(4593606)
x
```

```python
y = x.pivot(index ='Hora', columns ='Dia_semana')
y = y.reindex(columns=[('Cantidad_estimada', 'Lunes'), ('Cantidad_estimada', 'Martes'), ('Cantidad_estimada', 'Miercoles'), ('Cantidad_estimada', 'Jueves'), ('Cantidad_estimada', 'Viernes'), ('Cantidad_estimada', 'Sabado'), ('Cantidad_estimada',   'Domingo')])
```
